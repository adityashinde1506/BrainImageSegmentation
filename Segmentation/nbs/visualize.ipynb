```python
from dataloader import Dataset2d
from transforms import MinMaxScaler
```

```python
data_dir = Path("../data")
data_files = list((data_dir/"HGG").iterdir()) + \
             list((data_dir/"LGG").iterdir())

train_files = data_files[:200]
val_files = data_files[200:215]
test_files = data_files[215:]

train_dataset = Dataset2d(filenames=train_files, name="TrainSet", transform=[MinMaxScaler()])
```

```python
for i in range(10):
    
    idx = randint(100, 2000)
    
    img, lab = train_dataset[idx]
    
    fig = plotter.figure(figsize=(10,10))
    plotter.subplot(231)
    plotter.imshow(img[0, :, :], cmap="gray")
    plotter.subplot(232)
    plotter.imshow(img[1, :, :], cmap="gray")
    plotter.subplot(233)
    plotter.imshow(img[2, :, :], cmap="gray")
    plotter.subplot(234)
    plotter.imshow(img[3, :, :], cmap="gray")
    plotter.subplot(235)
    plotter.imshow(lab, cmap="gray")
    plotter.show()
    
    ones = numpy.sum(lab)
    print(f"Proportion = {ones/lab.ravel().shape[0]}")
    print(f"Max is {numpy.max(img)} and min is {numpy.min(img)}")
```
